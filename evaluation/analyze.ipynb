Analyze airbnb.com QUIC initials

Number of packets exchanged during loading airbnb.com

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
pageload_stats={
  "orig":      pd.read_csv('airbnb/orig.pageload.stats.csv'),
  "hacked":    pd.read_csv('airbnb/hacked.pageload.stats.csv'),
  "hacked_v2": pd.read_csv('airbnb/hacked_v2.pageload.stats.csv'),
}

quic_initial_stats={
  "orig":      pd.read_csv('airbnb/orig.quic_initial.stats.csv'),
  "hacked":    pd.read_csv('airbnb/hacked.quic_initial.stats.csv'),
  "hacked_v2": pd.read_csv('airbnb/hacked_v2.quic_initial.stats.csv'),
}

# Select relevant columns
# pageload_data = pd.Dataframe()
pageload_data_num_packets=pd.DataFrame()
pageload_data_num_packets["orig"]=pageload_stats["orig"]["num_packets"]
pageload_data_num_packets["hacked"]=pageload_stats["hacked"]["num_packets"]
pageload_data_num_packets["hacked_v2"]=pageload_stats["hacked_v2"]["num_packets"]

# Create the boxplot
plt.figure(figsize=(8, 6))

# Use a better color palette (e.g., "Set2" or "coolwarm")
# palette = sns.color_palette("Set3", 3)
sns.boxplot(data=pageload_data_num_packets)

# Customize the plot
plt.title('Statistics for the number of packets when visiting airbnb.com', fontsize=14)
plt.xlabel('DNS Types', fontsize=12)
plt.ylabel('Number of packets', fontsize=12)
plt.xticks(ticks=[0, 1, 2], labels=["Original DNS", "ApexDNS naive", "ApexDNS advanced"], fontsize=10)  # Custom x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a light grid for better readability

# Show the plot
plt.show()
```

Time to load airbnb.com

```python
# pageload_data = pd.Dataframe()
pageload_data_time=pd.DataFrame()
pageload_data_time["orig"]=pageload_stats["orig"]["time_to_load"]
pageload_data_time["hacked"]=pageload_stats["hacked"]["time_to_load"]
pageload_data_time["hacked_v2"]=pageload_stats["hacked_v2"]["time_to_load"]

# Create the boxplot
plt.figure(figsize=(8, 6))

# Use a better color palette (e.g., "Set2" or "coolwarm")
# palette = sns.color_palette("Set3", 3)
sns.boxplot(data=pageload_data_time)

# Customize the plot
plt.title('Statistics for the time required to load airbnb.com', fontsize=14)
plt.xlabel('DNS Types', fontsize=12)
plt.ylabel('Time to load [s]', fontsize=12)
plt.xticks(ticks=[0, 1, 2], labels=["Original DNS", "ApexDNS naive", "ApexDNS advanced"], fontsize=10)  # Custom x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a light grid for better readability

# Show the plot
plt.show()
```

QUIC Initial stats

i.e., when is the fist QUIC initial packet sent to airbnb.com
